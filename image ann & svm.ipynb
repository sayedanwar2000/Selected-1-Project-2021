```python
import numpy as np
import cv2
from tqdm import tqdm
import os
import matplotlib.pyplot as plt
import random
import keras
import tensorflow as tf
from keras.layers import Dense, Dropout,Flatten
import pickle
import numpy as np
import pandas as pd
from keras.layers import Conv2D
from keras.regularizers import l2
import tensorflow as tf
import os
import cv2
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import keras
IMG_SIZE = 50
```

```python
pickle_in = open("X_train_image.pickle","rb")
X_train = pickle.load(pickle_in)

pickle_in = open("y_train_image.pickle","rb")
y_train = pickle.load(pickle_in)

pickle_in = open("X_test_image.pickle","rb")
X_test = pickle.load(pickle_in)

pickle_in = open("y_test_image.pickle","rb")
y_test = pickle.load(pickle_in)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#ann = tf.keras.models.Sequential()
ann = tf.keras.models.Sequential()

ann.add(tf.keras.layers.Dropout(0.1))



#ann.add(tf.keras.layers.Dropout(0.2))

ann.add(tf.keras.layers.Dense(units=2500, activation='relu'))

ann.add(Dense(32, kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))

ann.add(tf.keras.layers.Dense(units=64, activation='relu'))

#ann.add(tf.keras.layers.Dense(units=10, activation='relu'))

ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

#checkpoint to hold the best weights found yet
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath="weights.h5", verbose=1, save_best_only=True)

historic =ann.fit(X_train, y_train, batch_size = 32, epochs = 300,validation_split=0.2, verbose=1, callbacks=[checkpoint])
```

```python
from keras.models import load_model
ann.save('ann.h5')
#ann = load_model("ann.h5")
```

```python
print(ann.evaluate(X_test, y_test))
print(ann.metrics_names)
```
